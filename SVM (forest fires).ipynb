```python
import pandas as pd 
import numpy as np 
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.svm import SVC 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix 
```

```python
data = pd.read_csv("forestfires.csv")
data
```

```python
pd.set_option('display.max_columns',None)
data 
```

```python
label_encoder=preprocessing.LabelEncoder()
```

```python
data['size_category']=label_encoder.fit_transform(data['size_category'])
data 
```

small = 1, large = 0

```python
data1=data.drop(['month','day'],axis=1)
data1 
```

```python
X = data1.iloc[:,0:28]
Y = data1['size_category']
```

```python
Y 
```

```python
X 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3) #30%
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

# Grid Search CV

# RBF

```python
clf = SVC() #SVC Classification
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10) #model
gsv.fit(X_train,Y_train) 
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 15, gamma = 50)
clf.fit(X_train , Y_train)
```

```python
Y_pred = clf.predict(X_test) 
Y_pred 
```

```python
acc = accuracy_score(Y_test, Y_pred) * 100
print("Accuracy =", acc)
```

```python
confusion_matrix(Y_test, Y_pred)
```

# Linear

```python
clf2 = SVC(kernel= "linear") 
clf2.fit(X_train , Y_train) 
```

```python
Y_pred2 = clf2.predict(X_test)
Y_pred2 
```

```python
acc2 = accuracy_score(Y_test, Y_pred2) * 100
print("Accuracy =", acc2) 
```

```python
confusion_matrix(Y_test, Y_pred2) 
```
